```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
```

```python
data= [1,4,2,8,5,7],[4,2,8,5,7,1],[0,8,5,7,1,4],[0,0,7,1,4,2],[0,0,0,4,2,8],[0,0,0,0,8,5],[0,0,0,0,0,7],[2,3,2,5,4,4],[3,2,5,4,4,2],[1,5,4,4,2,3],[1,1,4,2,3,2],[1,1,1,3,2,3],[1,1,1,1,5,4],[1,1,1,1,1,4]
print(data)
```

```python
img=np.array(data)
```

```python
def histogram(img):
    w=img.shape[0]
    h=img.shape[1]
    his=np.zeros(np.max(img)+1)
    for x in  range (w):
        for y in range (h):
            pixel=img[x,y]
            his[pixel]+=1
    return his

```

```python
def Vebieudohis(his):
    size=his.size
    X=np.zeros(size)
    X=np.linspace(0,size,size)
    plt.bar(X,his)
    plt.title('bieu do histogram')
    plt.xlabel('gia tri muc xam')
    plt.ylabel('so diem cung gia tri muc xam')
    plt.show()
```

```python
his=histogram(img)
print(his)
print(np.cumsum(his))
```

```python
def can_bang_his(his,new_level):
    w=img.shape[0]
    h=img.shape[1]
    m_n=w*h
    TB=m_n/new_level
    t_g=np.cumsum(his)
    f_g=lambda z: max(0,round(z/TB-1))
    his_equal=np.zeros(t_g.size)
    for i in range(t_g.size):
        his_equal[i]=f_g(t_g[i])
    return his_equal
```

```python
his_equal=can_bang_his(his,6)
```
