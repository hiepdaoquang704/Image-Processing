```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img=cv.imread('D:\imageprocessing\geometry.jpg',0)
edges=cv.Canny(img,50,150,apertureSize=3)
lines=cv.HoughLines(edges,rho=1,theta=np.pi/180,threshold=110)
for line in lines:
    rho,theta=line[0]
    a=np.cos(theta)
    b=np.sin(theta)
    x0=a*rho
    y0=b*rho
    x1=int(x0+1000*(-b))
    y1=int(y0+1000*(a))
    x2=int(x0-1000*(-b))
    y2=int(y0-1000*a)
    cv.line(img,(x1,y1),(x2,y2),(0),3)
```

```python

plt.imshow(img,cmap='gray')
```

```python
img=cv.imread('D:\imageprocessing\geometry.jpg',0)
edges=cv.Canny(img,50,150,apertureSize=3)
lines=cv.HoughLinesP(edges,2,np.pi/180,100,minLineLength=70,maxLineGap=50)
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv.line(img,(x1,y1),(x2,y2),(0,255,0),4)
plt.imshow(img,cmap='gray')
```
