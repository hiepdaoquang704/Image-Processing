```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img= cv.imread('D:\imageprocessing\geometry.jpg',0)
img=cv.medianBlur(img,5)
cricles=cv.HoughCircles(img,cv.HOUGH_GRADIENT,  10,40,param1=180,param2=170,minRadius=70,maxRadius=110)
cricles=np.uint16(np.around(cricles))
for i in cricles[0,:]:
    cv.circle(img,(i[0],i[1]),i[2],(0,255,0),2)
    cv.circle(img,(i[0],i[1]),2,(0,0,255),3)
plt.imshow(img,cmap='gray')
```
